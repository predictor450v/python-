```python
import pandas as pd
df_sales = pd.read_excel('appliance_sales_100_rows.xlsx')
df_sales.head()
```

```python
# Split Year_Quarter into sortable parts
df_sales[["Year", "Q"]] = df_sales["Year_Quarter"].str.split("-", expand=True)
df_sales["Q_num"] = df_sales["Q"].str.replace("Q", "").astype(int)

# Sort by time
df_sales = df_sales.sort_values(["Year", "Q_num"])
```

```python
from matplotlib import pyplot as plt

plt.Figure(figsize=(12,4))

plt.plot(df_sales["Year_Quarter"],df_sales["Fridge"], color ="blue")
plt.plot(df_sales["Year_Quarter"],df_sales["Dishwasher"],color ="red")
plt.plot(df_sales["Year_Quarter"],df_sales["Washing Machine"],color="black")



plt.xlabel("Year-Quarter")
plt.ylabel("Sales")
plt.title("Appliance Sales Over Time")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
