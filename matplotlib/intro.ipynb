```python
import pandas as pd
df_sales = pd.read_excel('appliance_sales_100_rows.xlsx')
df_sales.head()
```

```python
# Split Year_Quarter into sortable parts
df_sales[["Year", "Q"]] = df_sales["Year_Quarter"].str.split("-", expand=True)
df_sales["Q_num"] = df_sales["Q"].str.replace("Q", "").astype(int)

# Sort by time
df_sales = df_sales.sort_values(["Year", "Q_num"])
```

```python
from matplotlib import pyplot as plt

plt.Figure(figsize=(16,7))

plt.plot(df_sales["Year_Quarter"],df_sales["Fridge"], color ="blue",label="Fridge")
plt.plot(df_sales["Year_Quarter"],df_sales["Dishwasher"],color ="red",label="Dsihwasher")
plt.plot(df_sales["Year_Quarter"],df_sales["Washing Machine"],color="black",label="Washing Machine")


plt.legend()
plt.xlabel("Year-Quarter")
plt.ylabel("Sales")
plt.title("Appliance Sales Over Time")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.scatter(range(len(df_sales)), df_sales["Fridge"], label="Fridge")
plt.scatter(range(len(df_sales)), df_sales["Dishwasher"], label="Dishwasher")
plt.scatter(range(len(df_sales)), df_sales["Washing Machine"], label="Washing Machine")
plt.legend()
plt.show()

```
