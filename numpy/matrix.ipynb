Matrix method using numpy

```python
import numpy as np
```

```python

matrix_quater1 =np.array([[200,220,250],    #product1
                          [150,180,210],    #product2
                          [300,330,360]    #product3
                          ])
matrix_quater1   # sells of quater 1    
```

```python
matrix_quater2 =np.array([[209,231,259],  #product1
                          [155,192,222],  #product2
                          [310,340,375]  #product3
                          ])
matrix_quater2 #sells of quater 2
```

```python
#total sells of both quater 
total =(matrix_quater1+matrix_quater2)
total
```

```python
#improvmnt in q2
matrix_quater2-matrix_quater1
```

```python
#percentage of sells increase
(matrix_quater2-matrix_quater1)*100/matrix_quater1
```

```python
prices = np.array([
    [200,300,402],
    [300,212,453],
    [234,324,53]
])
q1_revenue = matrix_quater1*prices
q2_revenue = matrix_quater2*prices
```

```python
q1_revenue
```

```python
q2_revenue
```

```python
#discount 
dis_q1 = q1_revenue*0.2
dis_q2 = q2_revenue*0.2
dis_q1,dis_q2
```

```python
#net revenue = revenue - discount for both quater
net_revenue_q1 = q1_revenue - dis_q1
net_revenue_q2 = q2_revenue - dis_q2
net_revenue_q1,net_revenue_q2
```

```python
#total net revenue
tnr=np.sum(net_revenue_q1)
```
