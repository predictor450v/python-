```python
# stack
c = np.array([[101,'ayush',230],[102,'subho',234],[103,'ayushman',530]])
d = np.array([[101,'25th jan'],[102,'25jan'],[103,'23jan']])
```

```python
# horizontal stack
f = np.hstack((c,d))
f
```

```python
# vertical stack
e = np.array([[101,'ayush',230],[102,'subho',234],[103,'ayushman',530]])
g = np.vstack((c,e))
g
```

```python
# split - horizontal and vertical
np.hsplit(g,[1])
```

```python
g1 , g2 =np.vsplit(g,[3])
```
