```python
df = pd.read_csv("weather_data.csv",parse_dates=["day"])
df
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace=True)
df
```

```python
df.fillna(0 ) #,inplace = True
```

```python
df
```

```python
df.fillna({
    'temperature': df.temperature.mean(),  # or median
    'windspeed':df.windspeed.mean(),
    'event': 'No Event'
})
```

```python
df
```

```python
df.fillna(method='ffill')#forward fill
```

```python
df.fillna(method='bfill')#backward fill
```

```python
df.fillna(method='bfill',axis ='columns')#backward fill
```

```python
df
```

```python
df.interpolate()
```

```python
df.dropna()
```

```python
df.dropna(how='all')
```
