```python
import csv
```

```python
def calculate_rating_status(data,indrustry=None):
    rating = []
    for row in data:
        if row[3]!='NULL' and (not indrustry or row[1]==indrustry):
            rating.append(float(row[3]))
    min_rating = min(rating)
    max_rating = max(rating)
    avg_rating = sum(rating)/len(rating)
    return max_rating , min_rating ,avg_rating

with open("movies.csv") as f:
    data = list(csv.reader(f))
    header = data[0]
    data = data[1:]

    max_rating,min_rating,avg_rating = calculate_rating_status(data)
    print(f"all records:min rating ={min_rating},max rating ={max_rating}, avg rating={avg_rating}")
```
