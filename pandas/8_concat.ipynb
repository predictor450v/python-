```python
usa_weather = pd.DataFrame({
    "city":['new york','chicago','orlando'],
    "temp":[23,33,13],
    "humidity":[180,50,68]
})
usa_weather
```

```python
df1 = pd.DataFrame({
    "city":['mumbai','kolkata','delhi','bangalore'],
    "humidity":[80,90,100,78]
})
df1
```

```python
df2= pd.DataFrame({
    "city":['mumbai','kolkata','bangalore'],
    "temp":[32 ,43 ,21]
})
df2
```

how concat works?

```python
concated_dataframe = pd.concat([df1,df2],axis = 1)
concated_dataframe
```

how marge works?

```python
marged_dataframe = pd.merge(df1,df2,on="city")
marged_dataframe
```

```python
marged_dataframe = pd.merge(df1,df2,on="city" ,how ="outer")
marged_dataframe
```
