```python
import pandas as pd
```

```python
india_weather = pd.DataFrame({
    "city":['mumbai','kolkata','delhi','bangalore'],
    "temp":[32 ,43 ,43 ,21],
    "humidity":[80,90,100,78]
})
india_weather
```

```python
usa_weather = pd.DataFrame({
    "city":['new york','chicago','orlando'],
    "temp":[23,33,13],
    "humidity":[180,50,68]
})
usa_weather
```

***marge***

```python
df = pd.concat([india_weather,usa_weather])
df
```

```python
df = pd.concat([india_weather,usa_weather],ignore_index=True)
df
```

```python
df = pd.concat([india_weather,usa_weather],keys =["india","us"])
df
```
